```python
def review_product(search):
    #it store all the results
    reviews={"name":[],"rating":[],"short_review":[],"brief_review":[]}

    #Main url for request
    headers=({'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/124.0.0.0 Safari/537.36','Accept-Language': 'en-US, en;q=0.5'})
    url="https://www.flipkart.com/search?q="+search
    response =requests.get(url,headers=headers)
    page_html=response.content
    page_html_bs= bs(page_html,'html')
    link_product=page_html_bs.findAll("div",class_="cPHDOP col-12-12") #In link_product all the list of product store
    del link_product[0:3] #Deleted  1-3 rows because it is worth less

    review_link=[] # it store all the link of review for all products
    lk=""

    #Here, I tried to store all link in review_link
    for i in link_product:
        try:
            url="https://www.flipkart.com"+i.div.div.div.a['href']
            product_req=requests.get(url,headers=HEADER)
            product_bs4=bs(product_req.text)
            a=product_bs4.find("div",class_="col pPAw9M")
            lk="https://www.flipkart.com"+a.find_all("a")[-1]['href']
            review_link.append(lk)
        except:
            pass
    
    #Here, I get requests from every link of review link
    for i in review_link:
        review_request=requests.get(i,headers=HEADER)
        review_bs4=bs(review_request.text)
        review_boxes=review_bs4.find_all("div",class_="cPHDOP col-12-12")
        
        del review_boxes[0:2] #Deleted worthless div's
        del review_boxes[-1]

        #Here, I stores all the Name pf customers,their reviews in reviews dict
        for i in review_boxes:
            try:
                name=i.div.div.div.find("div",class_="row gHqwa8").div.p.string
                rating=i.div.div.div.div.div.text
                st_re=i.div.div.div.div.p.string
                b_re=i.div.div.div.find("div",class_="").div.text
                reviews["name"].append(name)
                reviews["rating"].append(rating)
                reviews["short_review"].append(st_re)
                reviews["brief_review"].append(b_re)
            except:
                pass
    #I converted all data into data frames for better understanding......   
    review_df=pd.DataFrame(reviews)
    return review_df
```
