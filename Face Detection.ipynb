```python
#this function is used to capture picture and it returns to the detection fuction related to the user choice
def capture(object):
    pic=cv2.VideoCapture(0)
    result, picture=pic.read()
    if object=='face':
        return faceDetection(picture)
    if object=='eye':
        return eyeDetection(picture)
    if object=='fullbody':
        return bodyDetection(picture)
```

```python
#this function is used to display example picture and it returns to the detection fuction related to the user choice
def example(object):
    if object=='face':
        picture=cv2.imread('C:\\Users\\ASUS\\Downloads\\detections\\people1.jpg')
        return faceDetection(picture)
    if object=='eye':
        picture=cv2.imread('C:\\Users\\ASUS\\Downloads\\detections\\people2.jpg')
        return eyeDetection(picture)
    if object=='fullbody':
        picture=cv2.imread('C:\\Users\\ASUS\\Downloads\\detections\\people3.jpg')
        return bodyDetection(picture)
```

```python
#This is face detection function, where is the .xml file used to detect face. from the function image come and it display the result
def faceDetection(picture):
    picture=cv2.resize(picture,(800,600))
    face_detector=cv2.CascadeClassifier('C:\\Users\\ASUS\\Downloads\\detections\\haarcascade_frontalface_default.xml')
    detect1=face_detector.detectMultiScale(picture,scaleFactor = 1.2,minNeighbors=7,minSize=(20,20), maxSize=(100,100))
    display(detect1)
    for (x,y,h,w) in detect1:
        cv2.rectangle(picture, (x, y), (x + h, y + w), (0,255,255), 3)
    cv2.imshow('',picture)
    cv2.waitKey(5000)
    cv2.destroyAllWindows()
```

```python
#This is eye detection function, where is the .xml file used to detect eye. from the function image come and it display the result
def eyeDetection(picture):
    picture=cv2.resize(picture,(800,600))
    eye_detector=cv2.CascadeClassifier('C:\\Users\\ASUS\\Downloads\\detections\\haarcascade_eye.xml')
    detect1=eye_detector.detectMultiScale(picture,scaleFactor=1.1,minNeighbors=10, maxSize=(60,60))
    display(detect1)
    for (x,y,h,w) in detect1:
        cv2.rectangle(picture, (x, y), (x + h, y + w), (0,255,255), 2)
    cv2.imshow('',picture)
    cv2.waitKey(5000)
    cv2.destroyAllWindows()
```

```python
#This is body detection function, where is the .xml file used to detect body. from the function image come and it display the result
def bodyDetection(picture):
    picture=cv2.resize(picture,(800,600))
    full_detector=cv2.CascadeClassifier('C:\\Users\\ASUS\\Downloads\\detections\\fullbody.xml')
    detect1=full_detector.detectMultiScale(picture,scaleFactor=1.05,minNeighbors=5,minSize=(50,50))
    display(detect1)
    for (x,y,h,w) in detect1:
        cv2.rectangle(picture, (x, y), (x + h, y + w), (0,255,255), 5)
    cv2.imshow('',picture)
    cv2.waitKey(5000)
    cv2.destroyAllWindows()
```
