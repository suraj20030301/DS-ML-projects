# Iris Dataset Analysis

In this notebook, we will explore the famous Iris dataset. This dataset is often used as a beginner's dataset for machine learning and data visualization tasks.

## Data Description

The Iris dataset consists of 150 samples of iris flowers, each described by four features:

- *Sepal length* (in cm)
- *Sepal width* (in cm)
- *Petal length* (in cm)
- *Petal width* (in cm)

There are three species of iris flowers in the dataset:
- *Iris setosa*
- *Iris versicolor*
- *Iris virginica*

Each species has 50 samples.

## Deploying the Models

1. Logistic Regression

```python
from sklearn.preprocessing import LabelEncoder
model_lc=LabelEncoder()
df['Species']=model_lc.fit_transform(df['Species'])
df
```

```python
x=df.iloc[:,:-1].values
x
```

```python
y=df.iloc[:,-1].values
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

```python
from sklearn.linear_model import LogisticRegression
model_lr=LogisticRegression()
model_lr.fit(x_train,y_train)
model_lr
```

```python
model_lr.score(x_train,y_train)
```

```python
y_pred=model_lr.predict(x_test)
y_pred
```
